```python
a= int(input("enter 1st no : "))
b= int(input("enter 2nd no : "))
a1=[]
b1=[]
for i in range(1,a+1):
    if a%i==0:
        a1.append(i)
for j in range(1,b+1):
    if b%j==0:
        b1.append(j)
a1=set(a1)
b1=set(b1)
if (a1&b1):
    l=list(a1&b1)
    
    
r=len(l)-1
r=l[r]
r1=a*b


ans = r1//r
print(f"{ans} is LCM of {a} , {b}")
```
